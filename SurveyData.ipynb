```python
# logic to get Usage Flag

def checkUsageMax(row):
    UsageFlagL = str(row.UsageFlag).split(",")        
    UsageFlagMax = max(UsageFlagL)
    return UsageFlagMax


    
def iterrows_impl(df):
    return pd.Series(
        checkUsageMax(row)     
        for row in df.itertuples()
    )
  

Comp5_Concat['UsageFlagMax'] = pd.Series(iterrows_impl(Comp5_Concat))

Comp5_Concat
```
